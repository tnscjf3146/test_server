## Model

### Model이란?
데이터베이스와 Python 클래스(객체)로 추상화된 형태로 상호작용
<br>
<br>

개발자가 데이터베이스에 대한 깊은 지식 없이도 쉽게 데이터 관리 가능<br>
유지보수 및 확장성 증대
<br>
<br>

- user.py : 사용자 요청의 시작점
- views.py : 요청을 처리하고 models.py를 통해 데이터를 다룸
- models.py : 데이터베이스를 정의하고, 상호작용
- templates : views.py에서 받은 데이터를 사용자에게 보여줄 화면을 구성

<br>
<br>

### Model class
DB의 테이블을 정의/설계도<br>
데이터를 조작할 수 있는 기능들을 제공

```python
# articles/models.py
from django.db import models

# Create your models here.
class Article(models.Model):
    title = models.CharField(max_length=10)
    content = models.TextField()
    created_at = models.DateTimeField(auto_now_add=True)
    updated_at = models.DateTimeField(auto_now=True)
```

### Model Field
DB 테이블의 필드(열) 정의<br>
데이터 타입 및 제약 조건 명시

<br>
<br>

### Field Types
데이터베이스에 저장될 "데이터의 종류"를 정의
- CharField() : 제한된 길이의 문자열을 저장 <br>필드의 최대 길이를 결정하는 max_length 옵션
- TextField() : 길이 제한이 없는 대용량 텍스트를 저장
- DateTimeField, TimeField, DateField
- IntegerField, FloatField
- FileField, ImageField
  
https://docs.djangoproject.com/en/5.2/ref/models/fields/#field-types

### Migrations
model 클래스의 변경사항(필드 생성, 수정 삭제 등)을 <br>
DB 에 최종 반영하는 방법

```python
# $ python manage.py makemigrations
# 모델 클래스 정의 후 마이그레이션 파일 생성

# $ python manage.py migrate
# 마이그레이션 파일을 DB에 반영
```
