## HTTP response status code
- - -
### HTTP
네트워크 상에서 데이터(리소스)를 주고 받기위한 약속

<br>

#### HTTP request methods란?
데이터에 대해 수행을 원하는 작업(행동)을 나타내는 것

<br>
<br>
<br>

#### "GET" Method
서버로부터 데이터를 요청하고 받아오는(조회) 데 사용<br>
- 데이터 전송
  - url의 쿼리 문자열을 통해 데이터를 전송
- 데이터 제한
  - url 길이 제한 대량 데이터 전송에는 적합 X
- 브라우저 히스토리
- 캐싱
  - 브라우저는 GET 요청의 응답을 로컬에 저장할 수 있음
  - 동일한 URL로 다시 요청할 때, 서버에 접속하지 않고 저장된 결과를 사용
  - 페이지 로딩 시간을 크게 단축

<br>
<br>
<br>

#### "POST" Method
서버에 데이터를 제출하여 리소스를 변경(생성, 수정, 삭제) 하는 데 사용
- 데이터 전송
  - HTTP Body를 통해 데이터를 전송
- 브라우저 히스토리
  - POST 요청은 브라우저 히스토리에 남지 않음

<br>
<br>

create 로직 수정

### HTTP response status code
서버가 클라이언트의 요청에 대한 처리 결과를 나타내는 3자리 숫자

<br>
<br>

#### 403 Forbidden
서버에 요청이 전달되었지만,<br>
권한 때문에 거절되었다는 의미

<br>
<br>
<br>

### CSRF
사이트 간 요청 위조
<br>
<br>
사용자가 자신의 의지와는 무관하게 공격자가 <br>의도한 행동(글쓰기, 정보 수정, 송금 등)을 특정 웹사이트에 요청하게 만드는 해킹 방식

### CSRF 토큰
Django의 안전장치

```python
<!-- templates/articles/new.html -->

<h1>NEW</h1>
<form action="{% url 'articles:create' %}" method="POST">
  {% csrf_token %}  # 토큰값도 같이 전송
  <div>
    <label for="title">Title: </label>
    <input type="text" name="title" id="title">
  </div>
  <div>
    <label for="content">Content: </label>
    <textarea name="content" id="content"></textarea>
  </div>
  <input type="submit">
</form>
<hr>
<a href="{% url 'articles:index' %}">[back]</a>
```

리소스의 변경을 요청하는 의미와 기술적 특성을 지니기에<dr>
DB에 조작을 가하는 요청은 반드시 인증 수단이 필요

### 필요한 views 함수
사용자 입력 데이터를 받을<br>
**edit**


```python
# articles/urls.py

urlpatterns = [
    # ...
    path('<int:pk>/edit/', views.edit, name='edit'),
]
```

```python
# articles/views.py

def edit(request, pk):
    article = Article.objects.get(pk=pk)
    context = {
        'article': article,
    }
    return render(request, 'articles/edit.html', context)
```

```python
<!-- articles/edit.html -->
<h1>EDIT</h1>
<form action="#" method="POST">
  {% csrf_token %}
  <div>
    <label for="title">Title: </label>
    <input type="text" name="title" id="title" value="{{ article.title }}">
  </div>
  <div>
    <label for="content">Content: </label>
    <textarea name="content" id="content">{{ article.content }}k/textarea>
  </div>
  <input type="submit">
</form>
<hr>
<a href="{% url 'articles:index' %}">[back]</a>
```

사용자가 입력한 요청 데이터를 받아 DB에 저장<br>
**update**

```python
# articles/urls.py

urlpatterns = [
    path('<int:pk>/update/', views.update, name='update'),
]
```

```python
# articles/views.py

def update(request, pk):
    article = Article.objects.get(pk=pk)
    article.title = request.POST.get('title')
    article.content = request.POST.get('content')
    article.save()
    return redirect('articles:detail', article.pk)    
```

```python
<!-- articles/edit.html -->
 
<h1>EDIT</h1>
<form action="{% url 'articles: update' article.pk %}" method="POST">
  {% csrf_token %}
  <div>
    <label for="title">Title: </label>
    <input type="text" name="title" id="title" value="{{ article.title }}">
  </div>
  <div>
    <label for="content">Content: </label>
    <textarea name="content" id="content">{{ article.content }}</textarea>
  </div>
  <input type="submit">
</form>
<hr>
<a href="{% url 'articles:index' %}">[back]</a>
```

### GET 과 POST

| 구분 | GET | POST|
| :---: | :---: | :---: |
|**데이터 전송 방식**| URL의 Query string parameter | HTTP body |
|**데이터 크기 제한**| 브라우저 제공 URL의 최대 길이 | 제한 없음 |
|**사용 목적**| 데이터 검색 및 조회 | 데이터 제출 및 변경 |

### 캐시
데이터나 정보를 임시로 저장하여 다시 요청할 때 빠르게 제공하는 저장공간
