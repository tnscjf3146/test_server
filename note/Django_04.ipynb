### CRUD
대부분의 소프트웨어가 가지는 기본적인 데이터 처리 기능인<br>
생성, 조회, 수정, 삭제를 묶어 이르는 말

#### 데이터 생성

```python
# 첫 번째 방법
# 빈 객체 생성
article = Article()

# title 정보 작성
article.title = 'first'

# content 정보 작성 
article.content = 'django!'

# 데이터베이스에 저장
article.save()

# 데이터베이스 Article 테이블에 저장된 모든 데이터 읽어서 확인
Article.objects.all()
```

```python
# 두 번째 방법
# 객체를 생성할 때 초기 값을 넣어서 생성
article = Article(title='second', content='django!')

# 데이터베이스에 저장
article.save()

# 데이터베이스 Article 테이블에 저장된 모든 데이터 읽어서 확인
Article.objects.all()
```

```python
# 세 번째 방법
# QuerySet API 인 create를 통해 한 번에 생성 및 저장
Article.objects.create(title='third', content='django!')

# 데이터베이스 Article 테이블에 저장된 모든 데이터 읽어서 확인
Article.objects.all()
```

### 데이터 조회 방법

```python
# all()
# 모델의 모든 데이터 조회
Article.objects.all()
```

```python
# filter
# content가 django!인 데이터 조회
Article.objects.filter(content='django!')

# title이 abc인 데이터 조회
Article.objects.filter(title='abc')

# title이 first인 데이터 조회
Article.objects.filter(title='first')
```

```python
# get
# 일치하는 데이터가 존재하는 경우
Article.objects.get(pk=1)
# 결과: <Article: Article object (1)>

# 일치하는 데이터가 없는 경우
Article.objects.get(pk=100)
# 결과: DoesNotExist: Article matching query does not exist.

# 일치하는 데이터가 여러 개 인 경우
Article.objects.get(content='django!')
# 결과: MultipleObjectsReturned: get() returned more than one Article -- it returned 2!
```

### 데이터 수정

```python
# update
# 1. 수정할 데이터를 가져온다.
article = Article.objects.get(pk=1)

# 2. 데이터를 수정한다.
article.title = 'byebye'

# 3. 저장한다.
article.save()
```

### 데이터 삭제

```python
# delete
# 1. 삭제할 데이터를 가져온다.
article = Article.objects.get(pk=1)

# 2. 삭제한다.
article.delete()
```

```python
# articles/urls.py
from django.urls import path
from . import views

urlpatterns = [
    path('', views.index, name='index'),
]
```

```python
# articles/views.py
from .models import Article

def index(request):
    articles = Article.objects.all()
    context = {
        'articles': articles,
    }

    return render(request, 'articles/index.html', context)
```

저장한 데이터 베이스 내의 데이터를<br>
for문으로 반복 출력

```python
<!-- templates/articles/index.html -->

<h1>Articles</h1>
<hr>
{% for article in articles %}
  <div>
    <p>글 번호: {{ article.pk }}</p>
    <p>글 제목: {{ article.title }}</p>
    <p>글 내용: {{ article.content }}</p>
  </div>
  <hr>
{% endfor %}
```

### Article.objects.filter(<span style='background-color:#fff5b1'>title</span>_ _<span style='background-color:#f5f0ff'>startswith</span>=<span style='background-color:#ffdce0'>'second'</span>)


Field Lookups는 모델의 필드 이름 뒤에 이중 밑줄을 붙이고, 원하는 조회 유형을 명시하는 방식으로 사용<br>
- exact / iexact
  - exact 대소문자를 구분하여 정확히 일치하는 값을 찾음
  - iexact 대소문자 구분 없이 정확히 일치하는 값을 찾음
- contains / icontains
  - contains: 문자열 내에 특정 값이 포함되어 있는값을 찾음(대소문자 구분)
  - icontains: 문자열 포함 여부를 대소문자 구분 없이 확인
- gt, gte, lt, lte
  - gt 초과
  - gte 이상
  - lt 미만
  - lte 이하
